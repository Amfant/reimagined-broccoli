```python
#Plot mass vs period
plt.plot(tab["pl_orbper"], tab["pl_bmassj"], "b.")

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Period [days]")
plt.ylabel("Mass [kg]")
#add solar system planets
ssperiods = [87.97, 224.7, 365.25, 687, 4332.59, 10759.22, 30687.15, 60190.03]
ssmasses = [0.00017, 0.00256, 0.00315, 0.00034, 1.00, 0.299, 0.046, 0.054]
ssnames = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Uranus", "Neptune"]

plt.scatter(ssperiods, ssmasses, color="red", label="Solar System Planets", zorder=5)
for i, name in enumerate(ssnames):
    plt.text(ssperiods[i], ssmasses[i], ssnames[i], fontsize=8, color="black")

plt.legend()
plt.show()
```
